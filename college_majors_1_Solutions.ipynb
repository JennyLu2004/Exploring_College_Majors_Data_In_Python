```python
# Read the csv file and store it in a variable
df = pd.read_csv("recent-grads.csv")
```

```python
# Getting to know our column names and and their data types
df.info()
```

### Data Exploration & Manipulation 

1. Find the top 5 major with the highest median salary 

```python
df.sort_values('Median', ascending = False).head(5)
```

2. Find the top 5 major with the lowest unemployment rate

```python
top_5_lowest_unemp = df.sort_values('Unemployment_rate', ascending = True).head(5)
top_5_lowest_unemp
```

3. Graph a barplot of the percentage of women in each major category. Which major category has the least share of women participation?

```python
plt.title("Percentage of women in each major category")
sns.barplot(data = df, x = 'ShareWomen', y = 'Major_category', color = 'dodgerblue')
plt.show()

print("Engineering has the least share of women participation")
```

4. Calculate the average median salary by major category, sort by descending order. Which major category has the lowest average median salary?

```python
df.groupby('Major_category')['Median'].mean().sort_values(ascending = False)
print('Psychology & Social Work has the lowest average median salary')
```

5. a. Which major category has the most amount of majors? How many majors does it have?

```python
df2 = df.groupby('Major_category')['Major'].count().sort_values(ascending = False)
df2
print('Engineering is the major category with the most major. It has 29 majors in total')
```

5. b. Create a countplot of major categories

```python
plt.title("Number of majors in each major category")
sns.countplot(data = df, y = 'Major_category', color = 'dodgerblue')
plt.show()
```

5. c. Get all the rows in the dataset where Major_category is Engineering

```python
df[df['Major_category']=='Engineering']
```

6. a. Find the correlation between Median Salary and Unemployment Rate

```python
df['Median'].corr(df['Unemployment_rate'])
```

6. b.  Visualize this correlation

```python
plt.title('Median Salary by Unemployment Rate')
sns.scatterplot(data = df, x = 'Median', y = 'Unemployment_rate')
plt.show()
```

7. Which major has the most students? The least students?

```python
# Major with the most student
df[df['Total'] == df['Total'].max()]

# Major with the least student 
df[df['Total'] == df['Total'].min()]
```

8. Create a boxplot to visualize Median Salary by Major Category

```python
plt.title('Median Salary by Major Category')
sns.boxplot(data = df, x = 'Median', y = 'Major_category')
plt.show()
```

9. Create a scatterplot to visualize the relationship between ShareWomen and the Median Salary

```python
plt.title("Percentage of Women by Median Salary")
sns.scatterplot(data = df, x = 'ShareWomen', y = 'Median')
plt.show()
```
